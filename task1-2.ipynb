```python
# coding:utf8
import tushare as ts
import pandas as pd
```

```python
df = ts.get_stock_basics()
stock = df.head(10).copy()
stock
```

```python
stock['riseDays'] = 0
stock
```

```python
code_list = stock.index.tolist()
code_list
```

```python
for code in code_list:
    stock.loc[(code, 'riseDays')] = 0
    price_change = ts.get_hist_data(code).price_change.tolist()
    for price_change_item in price_change:
        if price_change_item > 0:
            stock.loc[(code, 'riseDays')] += 1
        else:
            break
stock
```

```python
stock.to_csv('stock_risingDays.csv', encoding='utf_8_sig')
```
